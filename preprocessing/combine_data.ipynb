```python
import re
import pandas as pd
```

```python
raw_url1 = 'https://raw.githubusercontent.com/SendShieldLaskarAI/SendShield-Machine-Learning/25244aabf6d19daefe3d6d0c35ee4189ba4628e8/data/bully%20(1).csv'
raw_url2 = 'https://raw.githubusercontent.com/SendShieldLaskarAI/SendShield-Machine-Learning/8466386ac7875996ac13f778d588d58145079d9a/data/bully1.old.csv'

try:
    data1 = pd.read_csv(raw_url1, on_bad_lines='skip', encoding='utf-8')
    print("Successfully loaded data1")
    print(f"Shape: {data1.shape}")
except Exception as e:
    print(f"Error loading data1: {e}")
    data1 = None

try:
    data2 = pd.read_csv(raw_url2, on_bad_lines='skip', encoding='utf-8')
    print("Successfully loaded data2")
    print(f"Shape: {data2.shape}")
except Exception as e:
    print(f"Error loading data2: {e}")
    data2 = None
```
